```python
x = np.array([1, 7, 20, 50, 79])
y = np.array([10, 19, 30, 35, 51])
np.polyfit(np.log(x), y, 1)

# y ≈ 8.46 log(x) + 6.62

plt.figure()
plt.plot(x, y, 'ko', label="Original Noised Data")
plt.plot(x, 8.46 * np.log(x) + 6.62, 'r-', label="Fitted Curve")
plt.legend()
plt.show()
```
