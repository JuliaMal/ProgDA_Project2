```python
# Graph using numpy.linspace()  
# finding Skewness 
  
from scipy.stats import skew 
import numpy as np  
import pylab as p  
  
x1 = np.linspace( 3.5, 10, 20000 ) 
y1 = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(x1)**2  ) 
  
p.plot(x1, y1, '*') 
  
print( '\nSkewness for data : ', skew(y1))
```

```python
mu, sigma = 1, 1.0 # mean and standard deviation
s = np.random.lognormal(mu, sigma, 20000)
```

```python
s
```

```python
np.amin(s)
```

```python
np.amax(s)
```

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 100, density=True, align='mid')
```

```python
x = np.linspace(min(bins), max(bins), 20000)
pdf = (np.exp(-(np.log(x) - mu)**2 / (2 * sigma**2))
       / (x * sigma * np.sqrt(2 * np.pi)))
```

```python
plt.plot(x, pdf, linewidth=2, color='r')
plt.axis('tight')
plt.show()
```
